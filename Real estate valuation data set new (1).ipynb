# Visualisasi Data

```python
#harga tertinggi 2012-2013 (Afina)
harga_tertinggi = df["HousePrice"].max()
print(harga_tertinggi)
```

```python
#harga terendah 2012-2013 (Afina)
harga_terendah = df["HousePrice"].min()
print(harga_terendah)
```

```python
#antara umur rumah dengan harga (Dedi)
abs(df.corr())['HousePrice'].sort_values(ascending=False)
```

```python
#antara jarak stasiun mrt dengan harga 
plt.figure(figsize=(10,5))
plt.scatter(x=df['DistanceToTheNearestMRTStation'], y=df['HousePrice'])
plt.xlabel('The distance to the nearest MRT station (unit: meter)')
plt.ylabel('House price of unit area')
plt.title('Real Estate Valuation')
plt.show()
```

```python
#antara umur rumah dengan harga 
plt.figure(figsize=(10,5))
plt.scatter(x=df['houseAge'], y=df['HousePrice'])
plt.xlabel('House Age (unit: Age)')
plt.ylabel('House price of unit area')
plt.title('Real Estate Valuation')
plt.show()
```
